```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the stock symbols array for the desired stocks
stock_symbols = ["TATAMOTORS.NS", "INFY.NS", "RELIANCE.NS","ITC.NS","HDFCBANK.NS"]
```

```python
# Downloading the historical stock price data over a period of 6 months
stock_data = yf.download(stock_symbols, period='6mo')['Close']
```

```python
#Calculating daily returns
daily_returns = stock_data / stock_data.shift(1) - 1
daily_returns = daily_returns.dropna()
```
